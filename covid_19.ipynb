Nhận xét tổng quát về bộ dữ liệu
1. Bộ dữ liệu có 10 cột và 206.
2. Dữ liệu có 2 cột bị khuyết dữ liệu là cột country_code và cột population
3. cột day, month, year, cases, deaths => thuộc kiểu dữ liệu int64
4. cột country, continent, country_code, population => thuộc kiểu dữ liệu object 
5. cột population => thuốc kiểu dữ liệu float64
6. case, deaths, population là những cột thống kế số điếm, số lượng không được <0

```python
data1 = data[(data['date']=='2020-04-12')]
```

```python
plt.hist(data1['cases'], bins=200)
plt.title("Phân bố số ca mắc mới")
plt.xlabel("số số ca mắc mới")
plt.ylabel("Số lượng quốc gia")
```

```python
data1.describe()
```

Nhận xét phân bố dữ liệu ngày 2020-04-12
1. Dữ liệu có ca mắc mới trung bình ~ 376.8 case
2. Dữ liệu có độ lệch chuẩn ~ 2150.8 lớn hơn nhiều so với giá trị trung bình
3. Quốc gia có case mắc bệnh nhỏ nhất là: 0 và lớn nhất là 28391 lệch sai số với giá trị trung bình và độ lệch chuẩn
4. Có 50% số quốc gia có số ca nhiễm mới nhỏ hơn 11.5
5. Có 75% số quốc gia có số ca nhiễm mới nhỏ hơn 80.25
Kết luận:
Dữ liệu phân bố không đồng điều giữa các quốc gia
