Nhận xét dữ liệu:
    * Bộ dữ liệu có 8 cột và 7381 dòng

    * Place, ProductName, UnName được lưu trữ với kiểu dữ liệu object mang thuộc tính định danh => thước đo nominal

    * ProductId,UmId,Month, Year được lưu trữ với kiểu dữ liệu int64 tuy nhiên dùng để định danh sản phẩm => thước đo nomial
    
    * Price kiểu dữ liệu được lưu trữ với kiểu float dùng để thể hiện giá của sản phẩm, có thể tính toán và đo lường => thước đo ratio


```python
data_price = data.groupby('ProductName')['Price'].mean().reset_index()
data_price  
```
