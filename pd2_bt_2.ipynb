```python
import pandas as pd
```

```python
data = pd.read_csv('OnlineRetail.csv', encoding='ISO-8859-1')
```

```python
data.info()
```

Nhận xét về dữ liệu:
- Cấu trúc:
  + Bộ dữ liệu có 8 cột và 541909 dòng
  + Các cột InvoiceNo, StockCode, Description, Country thuộc kiểu dữ liệu object
  + Các cột UnitPrice, CustomerID thuộc kiểu dữ liệu float64
  + Cột Quantity thuộc kiễu dữ liệu int64
  * Thuộc tính, thước đo: 
    + InvoiceNo, StockCode, Description, Country là thuốc tính định tính, có thước đo định danh
    + CustomerID tuy mang kiểu dữ liệu là float64 nhưng nó dùng để phân biệt giữ các thực thể khác nhau nên mang thuộc tính định danh và thước đo định danh
    + Quantity, UnitPrice mang thuộc tính định lượng và thước đo ratio

```python
data1 = data.pivot_table(index=['InvoiceNo','Description','Country'], values='Quantity',aggfunc='mean')
data1
```

```python
data2 =data.pivot_table(index=['CustomerID'],columns=['StockCode'], values='Quantity', aggfunc={'Quantity':[max,min]})
data2
```

```python
data3 = data.pivot_table(values=['Quantity','UnitPrice'],index=['StockCode'],columns=['Description'], aggfunc={'Quantity':'sum','UnitPrice':'mean'})
```

```python
data3
```

```python
data4 = data.pivot_table(values='Quantity',index='InvoiceDate',columns='Description',aggfunc={'Quantity':'sum'})
```
